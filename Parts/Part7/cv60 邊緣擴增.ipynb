```python
'''各式邊緣擴增法'''
import cv2
import numpy as np
import matplotlib.pyplot as plt


filename = 'data/s.jpg'
img1 = cv2.imread(filename)
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
a = 20 #單方向擴增像素數量
replicate = cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_REFLECT)
wrap = cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_CONSTANT,value=(255,255,0))
print('ORIGINAL'), plt.imshow(img1), plt.axis('off')
plt.show()
print('REPLICATE'), plt.imshow(replicate), plt.axis('off') #邊緣像素延伸
plt.show()
print('REFLECT'), plt.imshow(reflect), plt.axis('off') #鏡像延伸
plt.show()
print('WRAP'), plt.imshow(wrap), plt.axis('off') #重覆拼貼
plt.show()
print('CONSTANT'), plt.imshow(constant), plt.axis('off') #填色
plt.show()
```

```python
'''cv60: 每秒換一次背景色'''
from time import sleep
from IPython.display import clear_output
n = 0
while n < 8:
  n += 1 
  B = np.random.randint(0,255) #產生隨機的RGB值
  G = np.random.randint(0,255)
  R = np.random.randint(0,255)
  
  sleep(1) #延遲1秒
  constant= cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_CONSTANT,value=(R,G,B))
  clear_output() #清掉輸出的圖案
  plt.imshow(constant), print('RGB=',(R,G,B)), plt.axis('off')
  plt.show()
```
