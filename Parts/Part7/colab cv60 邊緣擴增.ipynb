```python
'''各式邊緣擴增法'''
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

pathname = '/content/gdrive/My Drive/Colab Notebooks/'
filename = 'data/s.jpg'
img1 = cv2.imread(pathname + filename)
a = 20 #單方向擴增像素數量
replicate = cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_REFLECT)
wrap = cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_CONSTANT,value=(255,0,255))
print('ORIGINAL'), cv2_imshow(img1)
print('REPLICATE'), cv2_imshow(replicate) #邊緣像素延伸
print('REFLECT'), cv2_imshow(reflect) #鏡像延伸
print('WRAP'), cv2_imshow(wrap) #重覆拼貼
print('CONSTANT'), cv2_imshow(constant) #填色
```

```python
'''每秒換一次背景色'''
from time import sleep
from IPython.display import clear_output
n = 0
while n < 8:
  n += 1 
  B = np.random.randint(0,255) #產生隨機的RGB值
  G = np.random.randint(0,255)
  R = np.random.randint(0,255)
  
  sleep(1) #延遲1秒
  constant= cv2.copyMakeBorder(img1,a,a,a,a,cv2.BORDER_CONSTANT,value=[B, G ,R])
  clear_output() #清掉輸出的圖案
  cv2_imshow(constant), print([B, G ,R])
```
